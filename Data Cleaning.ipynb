# Add some values to missing data

```python
df.head()
```

```python
#Fill it with 0
df_with_0 = df.fillna(0)
df_with_0.head()
```

```python
df['DURATION'].mean()
```

```python
#Add mean value into the missing data in the duration columns
df_with_mean = df.DURATION.fillna(df['DURATION'].mean())
df_with_mean
```

# Dropping null column

```python
df.head()
```

```python
df.shape
```

```python
# Remove rows that have null values
df_drop = df.dropna()
df_drop.shape
```
